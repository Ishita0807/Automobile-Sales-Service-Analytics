```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
from datetime import datetime, timedelta
from google.colab import files
```

```python

sales_df = pd.read_csv("/content/drive/MyDrive/automobile_sales.csv")
service_df = pd.read_csv("/content/drive/MyDrive/automobile_service.csv")
```

```python
sales_df = pd.read_csv("/content/drive/MyDrive/automobile_sales.csv")
service_df = pd.read_csv("/content/drive/MyDrive/automobile_service.csv")


sales_df['Date'] = pd.to_datetime(sales_df['Date'])
service_df['Date'] = pd.to_datetime(service_df['Date'])
```

**EXPLORATORY DATA ANALYSIS**

**Monthly Sales Revenue**


```python
sales_df['Month'] = sales_df['Date'].dt.to_period('M')
monthly_sales = sales_df.groupby('Month')['Price'].sum()

monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Revenue', ylabel='Revenue (â‚¹)', figsize=(10,5))
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

**Sales by Region**

```python
region_sales = sales_df.groupby('Region')['Price'].sum().sort_values(ascending=False)

region_sales.plot(kind='bar', color='teal', title='Sales by Region', ylabel='Revenue (â‚¹)')
plt.tight_layout()
plt.show()
```

 **Sales by Brand**

```python
brand_sales = sales_df.groupby('Brand')['Price'].sum().sort_values(ascending=False)

sns.barplot(x=brand_sales.index, y=brand_sales.values, palette='Set2', hue=brand_sales.index, legend=False)
plt.title("Sales by Brand")
plt.ylabel("Total Revenue (â‚¹)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Most Common Service Issues**

```python
issue_counts = service_df['Issue_Type'].value_counts()

sns.barplot(x=issue_counts.values, y=issue_counts.index, palette='coolwarm', hue=issue_counts.index, legend=False)
plt.title("Most Common Service Issues")
plt.xlabel("Number of Cases")
plt.tight_layout()
plt.show()
```

**Dealer Performance Summary**


```python
sales_dealer = sales_df.groupby('Dealer_Name')['Price'].sum()
satisfaction_dealer = service_df.groupby('Dealer_Name')['Customer_Satisfaction'].mean()

dealer_summary = pd.concat([sales_dealer, satisfaction_dealer], axis=1).reset_index()
dealer_summary.columns = ['Dealer_Name', 'Total_Sales', 'Avg_Satisfaction']
dealer_summary
```

**Conclusion:**

Based on the exploratory data analysis, the following conclusions can be drawn:

* **Monthly Sales Revenue:** Sales show significant fluctuations throughout the year, with peaks in December and October, and a dip in September.
* **Sales by Region:** The East region has the highest sales revenue, followed by the South, West, and North regions.
* **Sales by Brand:** Kia is the top-selling brand by revenue, followed by Tata, Hyundai, and Maruti.
* **Most Common Service Issues:** "Transmission Lag", "Engine Overheat", and "Brake Issue" are the most frequently reported service problems.
* **Resolution Time vs Customer Satisfaction:** There appears to be no strong linear correlation between resolution time and customer satisfaction based on the scatter plot.
* **Dealer Performance Summary:** Elite Motors has the highest total sales and the highest average customer satisfaction among all dealers. Metro Cars has the lowest average customer satisfaction.
